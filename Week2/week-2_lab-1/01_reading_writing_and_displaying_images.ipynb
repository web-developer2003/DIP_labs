```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

# Read image
img = cv2.imread("/content/drive/MyDrive/DIP/team2.jpg")

# Convert BGR  RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display using OpenCV-style (Colab safe)
cv2_imshow(img)

# Display using matplotlib
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Changing the color channel order (BGR  RGB)</div>


<div style="direction:rtl;text-align:right;font-family:Tahoma">
Method 1: Using NumPy</div>

```python
#method1
plt.imshow(img[:,:,::-1])
```

<div style="direction:rtl;text-align:right;font-family:Tahoma">
Method 2: Using OpenCV</div>

```python
#method2
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Modifying and manipulating pixels</div>


```python

import cv2

img = cv2.imread('/content/drive/MyDrive/DIP/team2.jpg')
img[200:500,300:600,:] = 255
plt.imshow(img[...,::-1])
```

```python
img[200:300,300:500,:] = (0, 0, 255)
plt.imshow(img[...,::-1])
```
