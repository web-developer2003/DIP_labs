## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Converting images to grayscale or Grayscaling </div>

#### Grayscaling is process by which an image is converted from a full color to shades of grey (black & white)

In OpenCV, many functions grayscale images before processing. This is done because it simplifies the image, acting almost as a noise reduction and increasing processing time as there is less information in the image.

### Let convert our color image to greyscale

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('/content/drive/MyDrive/DIP/team2.jpg')

# Convert BGR  RGB (for correct colors)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Show original
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

# Show grayscale
plt.subplot(1,2,2)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load directly as grayscale (fast method)
img = cv2.imread('/content/drive/MyDrive/DIP/team2.jpg', 0)

print("Image shape:", img.shape)

# Display
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```
