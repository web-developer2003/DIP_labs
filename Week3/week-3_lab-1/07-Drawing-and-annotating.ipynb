## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Drawing images and shapes in OpenCV</div>

````python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image (RGB/BGR)
image = np.zeros((512, 512, 3), np.uint8)
# (512, 512, 3): height × width × 3 color channels (BGR)
# np.uint8: 8-bit unsigned integer type (pixel values 0–255)

# Create a black & white (grayscale) image
image_bw = np.zeros((512, 512), np.uint8)
# Only 2D because grayscale has a single channel

# Display color image (convert BGR → RGB for matplotlib)
plt.imshow(image[..., ::-1])
plt.title("Black Color Image")

# Display grayscale image
plt.figure()
plt.imshow(image_bw, cmap='gray')
plt.title("Black & White Image")
plt.show()
````

## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Drawing a line</div>
cv2.line(image, starting cordinates, ending cordinates, color, thickness)

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Draw a diagonal line from top-left to bottom-right
cv2.line(image, (0, 0), (511, 511), (255, 127, 0), 5)
# (255, 127, 0) is the color in BGR (Blue, Green, Red)
# 5 is the thickness of the line in pixels

# Display using OpenCV window
cv2.imshow("Blue Line", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR to RGB for matplotlib
plt.title("Diagonal Line")
plt.show()
````

Draws a **line** on an image using OpenCV.

- **`image`**: target image to draw on  
- **`(0, 0)`**: starting point of the line (top-left corner)  
- **`(511, 511)`**: ending point of the line (bottom-right corner)  
- **`(255, 127, 0)`**: color of the line in **BGR** format  
  - `255` = Blue, `127` = Green, `0` = Red  
- **`5`**: thickness of the line in pixels


### How OpenCV Handles Different Image Types

OpenCV treats images differently depending on their **data type**. This affects how the image appears when displayed or saved.

1. **8-bit unsigned (`np.uint8`)**  
   - Each pixel can have a value from **0 to 255**.  
   - OpenCV **shows it exactly as it is**.  
   - Example: black = 0, white = 255.

2. **16-bit unsigned (`np.uint16`)** or **32-bit integer (`np.int32`)**  
   - Pixel values can be much larger than 255 (up to 65535 for 16-bit).  
   - OpenCV **divides all values by 256** to bring them into the 0–255 range.  
   - Example: a pixel value of 65535 becomes 65535 ÷ 256 ≈ 255.

3. **32-bit floating-point (`np.float32`)**  
   - Pixel values usually range from **0.0 to 1.0**.  
   - OpenCV **multiplies all values by 255** to display them correctly.  
   - Example: a pixel value of 0.5 becomes 0.5 × 255 ≈ 128.

> In short: OpenCV always expects pixel values in the **0–255 range** when displaying images. Depending on the data type, it automatically scales the values.


## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Drawing a rectangle</div>

cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Draw a rectangle
cv2.rectangle(image, (100, 100), (300, 250), (0, 0, 255), 5)
# (100, 100) → top-left corner
# (300, 250) → bottom-right corner
# (0, 0, 255) → red color in BGR
# 5 → thickness of the rectangle border

# Display using OpenCV
cv2.imshow("Rectangle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR → RGB
plt.title("Rectangle")
plt.show()
````

Draws a **rectangle** on an image using OpenCV.

- **`image`**: target image to draw on  
- **`(100, 100)`**: top-left corner of the rectangle  
- **`(300, 250)`**: bottom-right corner of the rectangle  
- **`(0, 0, 255)`**: color of the rectangle in **BGR** format  
  - `0` = Blue, `0` = Green, `255` = Red (this makes it a red rectangle)  
- **`5`**: thickness of the rectangle's border in pixels  

> To make the rectangle **filled**, set the thickness to `-1`.

## <div style="direction:ltr;text-align:left;font-family:Arial, Tahoma">Drawing an Ellipse</div>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
ellipse = np.zeros((512, 512, 3), np.uint8)

# Draw a filled ellipse
cv2.ellipse(ellipse, (250, 250), (250, 100), 45, 90, 360, (255, 255, 255), -1)
# (250, 250) → center of the ellipse
# (250, 100) → lengths of the major and minor axes
# 45 → rotation angle in degrees
# 90 → starting angle of the arc
# 360 → ending angle of the arc
# (255, 255, 255) → color in BGR (white)
# -1 → fill the ellipse completely

# Display using OpenCV
cv2.imshow("Ellipse", ellipse)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(ellipse[..., ::-1])  # Convert BGR → RGB
plt.title("Ellipse")
plt.show()
````

Draws a **filled ellipse** on an image using OpenCV.

- **`ellipse`**: target image or mask  
- **`(250, 250)`**: center of the ellipse (x, y coordinates)  
- **`(250, 100)`**: lengths of the major (250) and minor (100) axes  
- **`45`**: rotation angle of the ellipse in degrees  
- **`90`**: starting angle of the ellipse arc  
- **`360`**: ending angle of the ellipse arc  
- **`(255, 255, 255)`**: color of the ellipse in **BGR** format (white)  
- **`-1`**: fills the ellipse completely  

> This is commonly used to create **filled shapes or masks**.


## <div style="direction:ltr;text-align:left;font-family:Arial, Tahoma">Drawing a Polygon</div>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Define four points of the polygon
pts = np.array([[10, 50], [400, 50], [90, 200], [50, 500]], np.int32)

# Reshape points for cv2.polylines: required shape (-1, 1, 2)
pts = pts.reshape((-1, 1, 2))

# Draw the polygon
cv2.polylines(image, [pts], True, (0, 0, 255), 3)
# True → close the polygon
# (0, 0, 255) → red color in BGR
# 3 → thickness of the lines

# Display using OpenCV
cv2.imshow("Polygon", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR → RGB
plt.title("Polygon")
plt.show()
````

Creates an array of **points** for drawing shapes in OpenCV.

- **`pts`**: variable storing the points  
- **`[[10, 50], [400, 50], [90, 200], [50, 500]]`**: list of coordinates (x, y) for each vertex  
- **`np.int32`**: data type required by OpenCV for drawing functions  

> This array is often used with `cv2.polylines()` or `cv2.fillPoly()` to draw **polygons**.


### Drawing Polylines in OpenCV

OpenCV provides the **`cv2.polylines()`** function to draw **connected lines (polylines)** efficiently.

1. **Closed vs. Open Polylines**  
   - If the **third argument** (`isClosed`) is `False`, the function draws a series of lines **connecting the points** without closing the shape.  
   - If `isClosed` is `True`, the last point is connected back to the first, forming a **closed polygon**.

2. **Drawing Multiple Polylines at Once**  
   - You can pass a **list of point arrays**, each representing a polyline, to `cv2.polylines()`.  
   - OpenCV will draw all polylines **individually**, in one call.  

3. **Why Use `cv2.polylines()` Instead of `cv2.line()`**  
   - Calling `cv2.line()` for every segment is **slower and less efficient**.  
   - `cv2.polylines()` is **faster** when drawing multiple connected lines or polygons.  

> Reference: [OpenCV Drawing Functions Documentation](https://docs.opencv.org/3.1.0/dc/da5/tutorial_py_drawing_functions.html)


### Adding Text in OpenCV

You can draw text on an image using the **`cv2.putText()`** function:

```python
cv2.putText(image, 'Text to Display', bottom_left_corner, font, font_scale, color, thickness)


### Parameters for `cv2.putText()`

- **`image`**: the target image to draw text on  
- **`'Text to Display'`**: the string you want to write  
- **`bottom_left_corner`**: starting point of the text (x, y coordinates)  
- **`font`**: font type (see list below)  
- **`font_scale`**: size of the text  
- **`color`**: text color in **BGR** format  
- **`thickness`**: thickness of the text strokes  

#### Available Fonts in OpenCV:

- `FONT_HERSHEY_SIMPLEX`  
- `FONT_HERSHEY_PLAIN`  
- `FONT_HERSHEY_DUPLEX`  
- `FONT_HERSHEY_COMPLEX`  
- `FONT_HERSHEY_TRIPLEX`  
- `FONT_HERSHEY_COMPLEX_SMALL`  
- `FONT_HERSHEY_SCRIPT_SIMPLEX`  
- `FONT_HERSHEY_SCRIPT_COMPLEX`


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Add text to the image
cv2.putText(image, 'Hello World!', (75, 290), cv2.FONT_HERSHEY_COMPLEX, 2, (100, 170, 0), 3)
# 'Hello World!' → text to display
# (75, 290) → bottom-left corner of the text
# cv2.FONT_HERSHEY_COMPLEX → font type
# 2 → font scale (size)
# (100, 170, 0) → text color in BGR
# 3 → thickness of the text strokes

# Display using OpenCV
cv2.imshow("Hello World!", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR → RGB
plt.title("Hello World!")
plt.show()
````
