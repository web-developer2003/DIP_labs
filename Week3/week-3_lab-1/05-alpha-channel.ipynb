## Alpha Channel and 4-Channel Images

In digital imaging, a **4-channel image** typically consists of **Red, Green, Blue, and Alpha (RGBA)** channels. The first three channels define color, while the **alpha channel** controls transparency.

### What Is the Alpha Channel?
The **alpha channel** stores opacity information for each pixel:
- **0 (or 0%)** → fully transparent  
- **1 (or 100%)** → fully opaque  
- Values in between → partial transparency

This allows images to blend smoothly with backgrounds, create soft edges, shadows, and visual effects.

### RGB vs. RGBA
- **RGB (3 channels):** Color only, no transparency  
- **RGBA (4 channels):** Color + transparency via alpha

### Common Uses
- PNG images with transparent backgrounds  
- UI elements and icons  
- Image compositing and layering  
- Video games, animation, and visual effects

### Why 4-Channel Images Matter
Using an alpha channel gives designers and developers precise control over how images appear when layered or rendered, making visuals cleaner and more flexible across different backgrounds.

> In short: **RGB defines what color a pixel is, alpha defines how visible it is.**


```python
import cv2
import matplotlib.pyplot as plt
```

Example to read PNG image containing four channels (Red, Green, Blue, Alpha).

```python
img = cv2.imread("pngimage.png", cv2.IMREAD_UNCHANGED)
img.shape
```

```python
bgr_image = img[...,0:3]   # bgr_image: extracts the first three channels of the image (Blue, Green, Red).
alpha_image = img[...,-1]  # alpha_image: extracts the last channel, which represents the alpha (transparency) channel.

rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image); plt.title("RGB");
```

```python
plt.imshow(alpha_image, cmap='inferno'); plt.title("Alpha");
```
